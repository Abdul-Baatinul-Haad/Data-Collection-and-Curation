```python
#display a histogram
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (12, 6))
plt.hist(df["median_house_value"], bins = 11, range = (14999, 500001))
plt.title("Median House Value Distribution")
plt.ylabel("Frequency")
plt.xlabel("Median House Value")
plt.axvline(x = np.mean(df["median_house_value"]), color = 'r', label = 'Mean')
plt.axvline(x = np.median(df["median_house_value"]), color = 'y', label = 'Median')
plt.legend()
```

```python
#display a box plot
fig2 = plt.figure(figsize = (10, 8))
plt.boxplot(df["housing_median_age"], widths = 1)
plt.title("Housing Median Age Boxplot")
plt.ylabel("Housing Median Age")
plt.xticks([])
```
